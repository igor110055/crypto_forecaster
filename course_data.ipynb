```python
from course_loader.loader_tools import CourseLoader
import datetime as dt
```

```python
##Tickers

def plus_name(tickers, name='USD'):
    new_tickers = []
    for ticker in tickers:
        new_tickers.append(ticker + '-' + name)
    return new_tickers

crypto_names = ['BTC', 'ETH', 'USDT', 'LTC', 'XRP', 'DASH', 'SC']
crypto_tikers = plus_name(crypto_names)

market_index_tickers =['^GSPC', '^DJI'] # ^DJI - Index Dow Jones.S&P500 [https://ffin.ru/market/directory/indexes/]
```

```python
# Analysis band

start_band = dt.datetime(2016, 1, 1)
end_band = dt.datetime(2022, 1, 1)
```

```python
cryptos = CourseLoader(start_band, end_band, crypto_tikers)
markets = CourseLoader(start_band, end_band, market_index_tickers)
```

```python
#Table dict for every crypto
cryptos.table_for_every_tickers[crypto_tikers[0]]
```

```python
# Crypto combined table whith close price
cryptos.make_combined_table(metric='Close')
```

```python
# Index combined table whith close price
markets.make_combined_table(metric='Close')
```
